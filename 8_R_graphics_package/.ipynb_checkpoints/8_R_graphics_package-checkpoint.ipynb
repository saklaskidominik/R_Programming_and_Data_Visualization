# Zadanie 1

1. Dane Cars93 (Pakiet MASS).
   - Wykonaj wykres rozrzutu zmiennej Price (o≈õ X) i Max.Price (o≈õ Y) za pomocƒÖ czerwonych litych tr√≥jkƒÖt√≥w, dodaj tytu≈Ç ‚ÄûPrice relationship‚Äù
   - Dodaj do wykresu w postaci punkt√≥w zale≈ºno≈õƒá miƒôdzy Price (o≈õ X) i Min.Price (o≈õ Y) funkcja points(), kolor niebieski, punkty okrƒôgi.

2. Dane Cars93 (Pakiet MASS).
   - Wykonaj wykres wyglƒÖdajƒÖcy jak przyk≈Çad poni≈ºej. Zastosuj logarytmowanie osi X i Y drugiego wykresu: log="xy"

3. Dane Cars93 (Pakiet MASS). Umie≈õƒá 3 wykresy w jednej ramce:
   - wykres ko≈Çowy (pie) liczby cylindr√≥w. Przydatna bƒôdzie funkcja table()
   - wykres s≈Çupkowy (barplot) z prostopad≈Çymi pomiƒôdzysonymi o 50% etykietami. Przydatne bƒôdƒÖ parametry las(), cex.names()
   - histogram (hist) dla Price. Zmodyfikuj kolory.


```r
options(repr.plot.width = 16, repr.plot.height = 8) # Ustawienie szeroko≈õci okna wykresu

# 1.
data(Cars93)
plot(Cars93$Price, Cars93$Max.Price, 
     type = "p", pch=17, col="red", xlab = "Price", 
     ylab = "Max price") 
title("Price relationship") 
points(Cars93$Price, Cars93$Min.Price, col = "blue", pch = 16)

# 2.
data(Cars93)

par(mfrow = c(1,2))
plot(Cars93$Luggage.room, Cars93$RPM, pch = 4, xlab = "Luggage room", ylab = "RPM")
title("Original representation")
plot(Cars93$Luggage.room, Cars93$RPM, xlab = "log Luggage room", ylab = "log RPM", pch=19, col="red", log="xy")
title("Log-log plot")

# 3.
par(mfrow = c(1,3))
pie(table(Cars93$Cylinders), col = terrain.colors(6))
title("Cylinders")
barplot(table(Cars93$Cylinders), las = 3, cex.names = 0.5, col = topo.colors(6))  
title("Cylinders")
hist(Cars93$Price, col = rainbow(12)) 
```
