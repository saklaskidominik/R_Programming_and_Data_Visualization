1. Napisz pętlę while zaczynając od x=0. Pętla wyświetla liczby z krokiem 1 do wartości 20, ale omija numer 12.
2.) Utwórz pętlę while z instrukcją warunkową. W pierwszym kroku przypisz do obiektu i wartość 1. Dopóki i<=10, oraz jest podzielne przez 3: wyświetl w konsoli wartość i przemnożoną przez 3. Gdy i będzie podzielne przez 8: wyświetl w konsoli komunikat 'Koniec' oraz przerwij działanie pętli. Na koniec napisz i o 1.


```r
# 1. Pętla while
liczby_bez_12 <- 0
while(liczby_bez_12<20)
{
  liczby_bez_12 <- (liczby_bez_12 + 1)
if( liczby_bez_12 == 12)
{
  next
}
  print(liczby_bez_12)
} 

#2. Pętla while z instrukcją warunkową
i <- 1
while(i <= 10)
{ 
  if(i%%3==0)
  {
  print(i*3)
  }
if( i%%8==0)
  {
    print("KONIEC")
    break
  }
  i <- i+1
}
```

## Zadanie 4

1. Sprawdź `?sapply` a następnie stwórz wektor `v <- c(1,2,3,4)` i oblicz pierwiastek dla każdego elementu wektora.

2. Zdefiniuj macierz składającą się z 10 wierszy i 3 kolumn o wartościach od 1 do 30 ułożone kolumnami. Następnie zsumuj elementy w każdym wierszu.

3. Utwórz listę:
   - Item1
     1 2 3 4 5
   - Item2
     4 12 20 28 36
   - Item3
     1 3 5 7 9
   
   Następnie oblicz wartość średnią elementów dla poszczególnych obiektów.

4. Dla każdej ze zmiennych ilościowych z mtcars oblicz statystyki opisowe i zapisz jako elementy listy.


```r
# 1. Obliczenie pierwiastka dla wektora
# ?sapply
v <- c(1,2,3,4)
pierwiastek_v <- sapply(v, sqrt)
cat("Pierwiastki wektora v:", pierwiastek_v, "\n\n")

# 2. Sumowanie elementów w macierzy
suma_macierzy <- matrix(1:30, nrow=10, ncol=3, byrow=FALSE)
cat("Macierz wartości od 1 do 30:\n")
print(suma_macierzy)
zsumowane <- apply(suma_macierzy, 1, sum)
cat("Suma elementów w każdym wierszu macierzy:\n", zsumowane, "\n\n")

# 3. Utworzenie listy i obliczenie średniej z elementów
lista <- list(
  item1 = c(1,2,3,4,5),
  item2 = c(4,12,20,28,36),
  item3 = c(1,3,5,7,9)
)
cat("Lista z elementami:\n")
print(lista)
srednia_obiektów <- sapply(lista, mean)
cat("Średnia dla każdego elementu listy:\n", srednia_obiektów, "\n\n")

# 4. Obliczenie statystyk opisowych dla zmiennych ilościowych z mtcars
zmienne_ilosciowe <- mtcars
statystyki_opisowe <- lapply(zmienne_ilosciowe, summary)
cat("Statystyki opisowe dla zmiennych ilościowych z mtcars:\n")
print(statystyki_opisowe)
```

## Zadanie 5

Utwórz pętlę while sprawdzającą prędkość auta i wyświetlającą informacje o prędkości. Jeśli prędkość będzie wyższa niż 80 lub niższa niż 30, wtedy pętla powinna się zakończyć. Jeśli prędkość jest wyższa niż 48 to powinien pojawić się komunikat „zwolnij mocno”, a od prędkości odejmij 11, w innych przypadkach powinien pojawić się komunikat zwolnij, a prędkość jest redukowana o 6. Prędkość początkowa wynosi 64.

```r
speed <- 64
while(speed < 80 && speed > 30)
{ 
  if(speed > 48)
  {
    print("Twoja predkosc to")
    print(speed)
    print("ZWOLNIJ MOCNO!!!")
    speed <- speed - 11
  }
  else
  {
    print("Twoja predkosc to")
    print(speed)
    print("ZWOLNIJ!")
    speed <- speed - 6
  }
}
```
