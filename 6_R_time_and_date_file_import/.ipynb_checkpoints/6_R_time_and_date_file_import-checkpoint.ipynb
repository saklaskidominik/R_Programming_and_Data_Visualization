## Zadanie 1
Nie korzystaj z Lubridate wykonaj zadania:

1. Zapisz jako datę w odpowiednim formacie
   - 30/July/2020
   - 1/15/2020
   - 1 Jan 20

2. Oblicz różnicę w dniach między dwoma datami (sprawdź `?difftime`)
   - "2020-10-15" 
   - "2020/02/11" (miesiąc listopad)

3. Zapisz urodziny do jednej zmiennej `bdays`:
   - Tukey: 1915-06-16
   - Fisher: 1890-02-17
   - Cramer: 1893-09-25
   i wyświetl w jakim dniu tygodnia się urodzili

4. Zapisz do zmiennej `mydate` 2020-4-19 7:01:00 (POSIXct)

5. Jaka będzie data za 100 dni od 2020-11-10?

6. Do zmiennej `my_fav_dates` zapisz 2005-4-19 8:01:00, 2006-4-19 11:01:00, 2008-4-19 7:01:00, sprawdź jaką średnią i maksymalną różnicę pomiędzy datami (sprawdź `?diff`)

7. Oblicz jaką ilość sekund dzieli zmienne zawierające informacje o dacie i godzinie:
   - 2000-11-10 12:00:00 
   - 2020-11-10 14:00:00


```r
# 1. Formatowanie dat
str_date1 <- "30/Lipiec/2020"
date1 <- as.Date(str_date1, format = "%d/%B/%Y")
str(date1)

str_date2 <- "1/15/2020"
date2 <- as.Date(str_date2, format = "%m/%d/%Y")
str(date2)

str_date3 <- "1 Sty 20"
date3 <- as.Date(str_date3, format = "%d %b %y")
str(date3)

# 2. Różnica w dniach
time1 <- as.Date("2020-10-15", format="%Y-%m-%d")
time2 <- as.Date("2020/02/11", format = "%Y/%d/%m")
difftime(time2, time1, units = "days")

# 3. Urodziny i dni tygodnia
bdays <- data.frame(
  name = c("Tukey", "Fisher", "Cramer"),
  birth_date = as.Date(c("1915-06-16", "1890-02-17", "1893-09-25"))
)
bdays$week_days <- weekdays(bdays$birth_date)

cat("Dni tygodnia urodzin: ", bdays$week_days, "\n")

# 4. Data i czas POSIXct
mydate <- as.POSIXct("2020-4-19 7:01:00", format="%Y-%m-%d %H:%M:%S")
print(mydate)

# 5. Data za 100 dni
data_za_100_dni <- as.Date("2020-11-10") + 100
cat("Data za 100 dni: ")
print(data_za_100_dni)

# 6. Średnia i maksymalna różnica pomiędzy datami
my_fav_dates <- as.POSIXct(c("2005-4-19 8:01:00", "2006-4-19 11:01:00", "2008-4-19 7:01:00"), format="%Y-%m-%d %H:%M:%S")
diffs <- diff(my_fav_dates)
cat("Różnice dat: ", diffs, "dni \n")
cat("Maksymalna różnica: ", max(diffs), "dni \n")
cat("Średnia różnica: ")
print(mean(diffs))

# 7. Ilość sekund między datami
start_time <- as.POSIXct("2000-11-10 12:00:00", format="%Y-%m-%d %H:%M:%S")
end_time <- as.POSIXct("2020-11-10 14:00:00", format="%Y-%m-%d %H:%M:%S")
seconds_diff <- difftime(end_time, start_time, units="secs")
cat("Ilość sekund różnicy:", seconds_diff, "sekund \n")
```

```r
# 1. Data urodzin 
b_day_2 <- ymd("2002-03-27")
cat("Dzień urodzin to:", day(b_day_2), "\n")
cat("Miesiąc urodzin to:", month(b_day_2), "\n")
cat("Rok urodzin to:", year(b_day_2), "\n")
cat("Rok przestępny:", ifelse(leap_year(b_day_2), "Tak", "Nie"), "\n")
cat("Dzień roku urodzin to:", yday(b_day_2), "\n")
cat("Numer tygodnia urodzin to:", week(b_day_2), "\n")
cat("Dzień tygodnia urodzin to:", weekdays(b_day_2), "\n")

# 2. Wydruk wektora dat
wektor_dat <- c("30-2020-01", "15/2020/02")
cat("Wektor dat:")
print(wektor_dat)

# 3. Daty utworzone z wektora dat
date4 <- parse_date_time(wektor_dat, orders = c("d-y-m", "d/y/m"))
cat("Daty utworzone z wektora dat:")
print(date4)

# 4. Obecna data i godzina
class_d_h <- now()
cat("Obecna data i godzina:")
print(class_d_h)

# 5. Data i godzina w strefie czasowej Australii
class_d_h_a <- force_tz(class_d_h, tzone = "Australia/Sydney")
cat("Data i godzina w strefie czasowej Australii:")
print(class_d_h_a)

# 6. Data i godzina w strefie czasowej Nowego Jorku, bez zmiany godziny
class_d_h_f <- with_tz(class_d_h, tzone = "America/New_York")
cat("Data i godzina w strefie czasowej Nowego Jorku:")
print(class_d_h_f)

# 7. Data i godzina z dodanymi dwoma godzinami
dodanie_2_godziny <- interval(class_d_h, class_d_h + hours(2))
cat("Data i godzina z dodanymi dwoma godzinami:")
print(dodanie_2_godziny)

# 8. Data i godzina zaokrąglona do najbliższej pełnej godziny
class_d_h_round <- round_date(class_d_h, unit = "hour")
cat("Data i godzina zaokrąglona do najbliższej pełnej godziny:")
print(class_d_h_round)

# 9. Daty dla Auckland i Arizony z odpowiednimi strefami czasowymi
date_Auckland <- ymd_hms("2020-10-10 10:10:00", tz= "Pacific/Auckland")
date_Arizona <- ymd_hms("2020-10-10 10:10:00", tz = "US/Arizona")
cat("Data i godzina dla Auckland:")
print(date_Auckland)
cat("Data i godzina dla Arizony:")
print(date_Arizona)
roznica_godzin_strefy <- as.numeric(difftime(date_Arizona, date_Auckland, units = "hours"))
cat("Różnica godzin pomiędzy strefami wynosi:", roznica_godzin_strefy, "godzin\n")

# 10. Różnica dni między datą urodzenia a obecną datą
roznica_dni_urodziny <- as.numeric(difftime(class_d_h, b_day_2, units = "days"))
cat("Różnica dni między datą moich urodzin a obecną datą:", roznica_dni_urodziny, "dni\n")

# 11. Różne formaty zapisu daty 2021-11-10
data_1 <- ymd("2021-11-10")
data_2 <- ydm("2021-11-10")
data_3 <- mdy("11/10/2021")
data_4 <- ymd(20211110)
data_5 <- mdy("November 10th, 2021")
data_6 <- dmy("10th of November '21")
cat("Data 2021-11-10 w różnych formatach:")
print(data_1)
print(data_2)
print(data_3)
print(data_4)
print(data_5)
print(data_6)
```

```r
# 1. Wczytanie danych z pliku 'hotdogs.txt'
hotdogs_data <- read.delim("hotdogs.txt", header = TRUE, sep = "\t", quote = "\"", dec = ".", fill = TRUE, comment.char = "")
cat("Wczytane dane z pliku 'hotdogs.txt':\n")
hotdogs_data

# Podsumowanie danych
cat("Podsumowanie danych z pliku 'hotdogs.txt':\n")
summary(hotdogs_data)

# 2. Wyszukiwanie hot-doga z najmniejszą liczbą kalorii
min_kalorii <- hotdogs_data[which.min(hotdogs_data[, 3]), ]
cat("Hot-dog z najmniejszą liczbą kalorii:\n")
min_kalorii

# Wyszukiwanie hot-doga z najwyższą zawartością sodu
max_sodu <- hotdogs_data[which.max(hotdogs_data[, 2]), ]
cat("Hot-dog z najwyższą zawartością sodu:\n")
max_sodu
```
