## Zadanie 1

1. Utwórz sekwencje:
   - `seq(1:30)`
   - `seq(30:1)`
   - `seq(from=0, to=1, by=0.25)`
   - `seq(from=20, to=5, length.out=10)`

2. Utwórz liczby losowe z rozkładu normalnego:
   - `a <- rnorm(100, mean=0, sd=10)`
   - `hist(a)`

3. Utwórz powtórzenia:
   - `rep(1:4, 5)`

```r
# 1. Utwórz sekwencje:
seq1to30 <- seq(1:30)
cat("Sekwencja od 1 do 30: ", seq1to30, "\n")

seq30to1 <- seq(30:1)
cat("Sekwencja od 30 do 1: ", seq30to1, "\n")

seq0to1 <- seq(from=0, to=1, by=0.25)
cat("Sekwencja od 0 do 1 z krokiem 0.25: ", seq0to1, "\n")

seq20to5 <- seq(from=20, to=5, length.out=10)
cat("Sekwencja od 20 do 5, 10 elementów równomiernie rozłożonych: ", seq20to5, "\n")

# 2. Utwórz liczby losowe z rozkładu normalnego:
a <- rnorm(100, mean=0, sd=10)
cat("Pierwsze 6 wartości z wektora 'a' o rozkładzie normalnym N(0,10): ", head(a), "\n")
options(repr.plot.width = 14, repr.plot.height = 8) # Ustawienie szerokości okna wykresu
hist(a, main="Histogram liczb losowych z rozkładu normalnego")

# 3. Utwórz powtórzenia:
rep1to4 <- rep(1:4, 5)
cat("Powtórzenie sekwencji 1 do 4, pięciokrotnie: ", rep1to4, "\n")
```

4. Wygeneruj zmienną normal, która będzie zawierać 1000 obserwacji o średniej 10 i odchyleniu standardowym 1.
5. Wygeneruj zmienną `pow`, w której będzie przechowywany wektor zawierający cyfry od 1 do 10 powtórzone 3 razy.

```r
# 1. Sprawdź jakie argumenty może mieć funkcja seq.
cat("Argumenty funkcji seq: \n")
#?seq

# 2. Wygeneruj sekwencję od -6 do 6 z krokiem 2.
cat("Sekwencja od -6 do 6 z krokiem 2: ", seq(from=-6, to=6, by=2), "\n")

# 3. Wygeneruj sekwencję od -2 do 5, łącznie 15 obserwacji.
cat("Sekwencja od -2 do 5, łącznie 15 obserwacji: ", seq(from=-2, to=5, length.out=15), "\n")

# 4. Wygeneruj zmienną normal, która będzie zawierać 1000 obserwacji o średniej 10 i odchyleniu standardowym 1.
normal <- rnorm(1000, mean=10, sd=1)
cat("Wygenerowana zmienna normalna: pierwsze 6 wartości: ", head(normal), "\n")

# 5. Wygeneruj zmienną pow, w której będzie przechowywany wektor zawierający cyfry od 1 do 10 powtórzone 3 razy.
pow <- rep(1:10, 3)
cat("Wektor powtórzony 3 razy: ", pow, "\n")
```

## Zadanie 3

1. Przepisz zmienną i funkcję do konsoli i sprawdź wyniki:
   - `m1 <- matrix(1:9, ncol=3)`
   - `m2 <- matrix(1:9, nrow=3)`
   - `m3 <- matrix(1:9, nrow=3, byrow=TRUE)`
   - Jaką jest różnica między macierzami? Zapisz komentarz używając `#`

2. Inny sposób generowania macierzy:
   - `a <- c(9,10,11)`
   - `b <- c(6,9,12)`
   - `c <- c(16,18,20)`
   - `mab <- matrix(c(a,b,c), ncol=3, byrow=TRUE)`
   - `rownames(mab) <- c("a","b","c")`

3. Macierze obliczenia:
   - `m1k <- m1+2`
   - `m2k <- m2*3`

4. Wybieranie elementów; napisz komentarz używając `#` który element został zwrócony
   - `m1[2,3]`
   - `m1[2, ]`
   - `m1[,1]`


```r
# 1. Przepisz zmienną i funkcję do konsoli i sprawdź wyniki:
m1 <- matrix(1:9, ncol=3)
m2 <- matrix(1:9, nrow=3)
m3 <- matrix(1:9, nrow=3, byrow=TRUE)

cat("Macierz m1 (wypełnienie kolumnowe): \n")
print(m1)
cat("Macierz m2 (wypełnienie kolumnowe, ale zdefiniowane przez wiersze): \n")
print(m2)
cat("Macierz m3 (wypełnienie wierszowe): \n")
print(m3)
# Komentarz do różnicy między m1, m2, a m3:
cat("# m1 i m2 są wypełniane kolumnami, ale m2 jest zdefiniowane przez ilość wierszy; m3 jest wypełniana wierszami.\n")

# 2. Inny sposób generowania macierzy:
a <- c(9,10,11)
b <- c(6,9,12)
c <- c(16,18,20)
mabc <- matrix(c(a,b,c), ncol=3, byrow=TRUE)
rownames(mabc) <- c("a", "b", "c")

cat("Macierz mabc (wypełnienie wierszowe z nazwami wierszy): \n")
print(mabc)

# 3. Macierze obliczenia:
m1k <- m1 + 2
m2k <- m2 * 3

cat("Macierz m1k (m1 + 2): \n")
print(m1k)
cat("Macierz m2k (m2 * 3): \n")
print(m2k)

# 4. Wybieranie elementów; napisz komentarz używając `#` który element został zwrócony
cat("Element m1[2,3] - trzeci element drugiego wiersza macierzy m1: ", m1[2,3], "\n")
cat("Drugi wiersz macierzy m1: \n")
print(m1[2, ])
cat("Pierwsza kolumna macierzy m1: \n")
print(m1[,1])
```

## Zadanie 4

1. Utwórz wektor `w1` składający się z 16 liczb od 2 do 17.
2. Wygeneruj 2 macierze biorąc `w1` jako argument: `mw1` i `mw2` różniące się ułożeniem elementów.
3. Podziel macierz `mw1` przez 3.
4. Przemnóż macierz `mw1` przez `mw2`.
5. Sprawdź, które elementy powstałej w punkcie 4. macierzy są większe lub równe 20.
6. Napisz kod, który da następujący wynik:

   |               | US      | non-US |
   |---------------|---------|--------|
   | A New Hope    | 460.998 | 314.4  |
   | The Empire Strikes Back | 290.475 | 247.9  |
   | Return of the Jedi | 309.306 | 165.8  |

7. Wybierz z macierzy powstałej w pkt 6 ostatni wiersz danych.
8. Wybierz z macierzy powstałej w pkt 6 wartość `314.4`.
9. Wygeneruj macierz składającą się z 9 obserwacji z rozkładu normalnego o wartości średniej wynoszącej 6 i odchyleniu standardowym równym 12.


```r
# 1. Utwórz wektor w1 składający się z 16 liczb od 2 do 17.
w1 <- 2:17
cat("Wektor w1: ", w1, "\n")

# 2. Wygeneruj 2 macierze biorąc w1 jako argument: mw1 i mw2 różniące się ułożeniem elementów.
mw1 <- matrix(w1, nrow=8)
mw2 <- matrix(w1, ncol=8)
cat("Macierz mw1 (wierszami): \n")
print(mw1)
cat("Macierz mw2 (kolumnami): \n")
print(mw2)

# 3. Podziel macierz mw1 przez 3.
mw3 <- mw1 / 3
cat("Macierz mw3 (mw1/3): \n")
print(mw3)

# 4. Przemnóż macierz mw1 przez mw2.
mw4 <- mw1 %*% mw2
cat("Macierz mw4 (mw1 %*% mw2): \n")
print(mw4)

# 5. Sprawdź, które elementy powstałej w punkcie 4 macierzy są większe lub równe 20.
ele_wieksze <- mw4 >= 20
cat("Elementy macierzy mw4 >= 20: \n")
print(ele_wieksze)

# 6. Stworzenie i wypisanie tabeli danych filmów.
US <- c(460.998, 290.475, 309.306)
non_US <- c(314.4, 247.9, 165.8)
tabela <- matrix(c(US, non_US), ncol=2, byrow=FALSE)
rownames(tabela) <- c("A New Hope", "The Empire Strikes Back", "Return of the Jedi")
colnames(tabela) <- c("US", "non-US")
cat("Tabela danych filmów: \n")
print(tabela)

# 7. Wybierz z macierzy powstałej w pkt 6 ostatni wiersz danych.
cat("Ostatni wiersz tabeli: \n")
print(tabela[3, ])

# 8. Wybierz z macierzy powstałej w pkt 6 wartość 314.4.
cat("Wartość 314.4 z tabeli: ", tabela[1, 2], "\n")

# 9. Wygeneruj macierz składającą się z 9 obserwacji z rozkładu normalnego o wartości średniej wynoszącej 6 i odchyleniu standardowym równym 12.
macierz_normalna <- matrix(rnorm(9, mean=6, sd=12), nrow=3)
cat("Macierz 3x3 z rozkładu normalnego (średnia=6, sd=12): \n")
print(macierz_normalna)
```
